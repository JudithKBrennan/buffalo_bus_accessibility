```python
# Function to calculate average latitude
def calculate_avg_lat(row):
    latitudes = [row['top-left'][0], row['bottom-left'][0], row['bottom-right'][0], row['top-right'][0]]
    return sum(latitudes) / len(latitudes)

# Function to calculate average longitude
def calculate_avg_lon(row):
    longitudes = [row['top-left'][1], row['bottom-left'][1], row['bottom-right'][1], row['top-right'][1]]
    return sum(longitudes) / len(longitudes)

# Masten Park
# Convert dictionary to DataFrame
dfmp = pd.DataFrame.from_dict(mastenParkBlockBoundaries, orient='index', columns=['top-left', 'bottom-left', 'bottom-right', 'top-right'])

# Reset index to make 'ID' a column instead of index
dfmp.reset_index(inplace=True)

dfmp.rename(columns={'index': 'block_id'}, inplace=True)

# Apply the functions to each row to calculate average latitude and longitude separately
dfmp['avg_lat'] = dfmp.apply(calculate_avg_lat, axis=1)
dfmp['avg_lon'] = dfmp.apply(calculate_avg_lon, axis=1)
dfmp['neighborhood'] = 'Masten Park'

# Fruit Belt
# Convert dictionary to DataFrame
dffb = pd.DataFrame.from_dict(fruitBeltBlockBoundaries, orient='index', columns=['top-left', 'bottom-left', 'bottom-right', 'top-right'])

# Reset index to make 'ID' a column instead of index
dffb.reset_index(inplace=True)
dffb.rename(columns={'index': 'block_id'}, inplace=True)

# Apply the functions to each row to calculate average latitude and longitude separately
dffb['avg_lat'] = dffb.apply(calculate_avg_lat, axis=1)
dffb['avg_lon'] = dffb.apply(calculate_avg_lon, axis=1)
dffb['neighborhood'] = 'Fruit Belt'

# Allentown
# Convert dictionary to DataFrame
dfat = pd.DataFrame.from_dict(AllentownBlockBoundaries, orient='index', columns=['top-left', 'bottom-left', 'bottom-right', 'top-right'])

# Reset index to make 'ID' a column instead of index
dfat.reset_index(inplace=True)
dfat.rename(columns={'index': 'block_id'}, inplace=True)

# Apply the functions to each row to calculate average latitude and longitude separately
dfat['avg_lat'] = dfat.apply(calculate_avg_lat, axis=1)
dfat['avg_lon'] = dfat.apply(calculate_avg_lon, axis=1)
dfat['neighborhood'] = 'Allentown'

# BNMC
# Convert dictionary to DataFrame
dfbnmc = pd.DataFrame.from_dict(BNMCBlockBoundaries, orient='index', columns=['top-left', 'bottom-left', 'bottom-right', 'top-right'])

# Reset index to make 'ID' a column instead of index
dfbnmc.reset_index(inplace=True)

dfbnmc.rename(columns={'index': 'block_id'}, inplace=True)

# Apply the functions to each row to calculate average latitude and longitude separately
dfbnmc['lat'] = dfbnmc.apply(calculate_avg_lat, axis=1)
dfbnmc['lon'] = dfbnmc.apply(calculate_avg_lon, axis=1)
dfbnmc['neighborhood'] = 'BNMC'
dfbnmc['point_id'] = range(1, len(dfbnmc) + 1)
```
